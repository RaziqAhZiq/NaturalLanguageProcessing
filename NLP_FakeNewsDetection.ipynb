## Evaluation Methodology
Evaluating the performance of machine learning models in fake news detection requires a comprehensive set of metrics to ensure accurate and meaningful comparisons. This project will employ Accuracy, Precision, Recall, and F1-Score as the primary metrics for assessing model performance, following closely the evaluation approach outlined in (Nurhasanah et al., 2022).

To establish a baseline performance, we use a DummyClassifier from sklearn with the "most_frequent" strategy. This strategy predicts the most frequent class in the training data for all instances. The rationale behind using the most frequent strategy is that it provides a simple and intuitive benchmark. If the majority of our news articles are real (or fake), the most frequent class prediction can act as a naive baseline to compare against our more sophisticated models.

**1. Accuracy:**
Accuracy is the proportion of correctly predicted instances (both true positives and true negatives) to the total number of instances. It provides a broad measure of model performance but can be misleading if the dataset is imbalanced.


 

**2. Precision:**
Precision measures the proportion of true positive predictions among all positive predictions. This metric is important in scenarios where false positives carry a high cost, such as falsely labeling legitimate news as fake.


 

**3. Recall:**
Recall, or sensitivity, is the proportion of true positive predictions among all actual positive instances. This metric is crucial for ensuring that fake news is accurately identified and not missed.


 

**4. F1-Score:**
The F1-Score is the harmonic mean of Precision and Recall. It provides a single metric that balances both Precision and Recall, particularly useful in dealing with imbalanced datasets.


 

### Application of Metrics
**Comparative Analysis:**
To compare statistical and embedding-based models, these metrics will be calculated for each model. The evaluation process involves:

1. **Training and Testing:** The embedding-based model will be trained on a training set and evaluated on a test set.
2. **Confusion Matrix:** Generating a confusion matrix for each model to determine TP, TN, FP, and FN counts.
3. **Precision-Recall Curve:** Plotting the precision-recall curve for each model to evaluate the trade-off between precision and recall.
4. **ROC and AUC Curve:** Plotting the ROC (Receiver Operating Characteristic) curve and calculating the AUC (Area Under the Curve) for each model to assess the discriminative ability.
5. **Comparison:** Analyzing these metrics to determine which model performs better in detecting fake news.

**Interpretation:**
- **Accuracy** provides an overall correctness measure but requires careful interpretation in the presence of class imbalance.
- **Precision** is critical for minimizing false positives, ensuring that articles identified as fake are genuinely fake.
- **Recall** is essential for capturing all instances of fake news, reducing the risk of missing actual fake articles.
- **F1-Score** offers a balanced metric that accounts for both Precision and Recall, giving a comprehensive view of model performance.

By employing these metrics, this project can systematically evaluate the efficacy of statistical and embedding-based models in fake news detection.

The final step of preprocessing involves cleaning and normalizing the text data to prepare it for modeling. A custom function, preprocess_text, is defined to handle this task:

**Remove special characters and punctuation:**
- text = re.sub(r'\W', ' ', str(text)): This removes any non-word characters (such as punctuation marks) and replaces them with a space.

**Replace multiple spaces with a single space:**
- text = re.sub(r'\s+', ' ', text, flags=re.I): This ensures that multiple consecutive spaces are replaced by a single space, making the text more uniform.

**Remove single characters surrounded by spaces:**
- text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text): This removes single characters that are surrounded by spaces, which are often not meaningful in text analysis.

**Remove numbers:**
- text = re.sub(r'\d+', '', text): This removes any numeric digits from the text, which might not be useful for text-based models.

**Convert text to lowercase:**
- text = text.lower(): This converts all characters in the text to lowercase to ensure uniformity and reduce case-sensitivity issues.

**Remove stopwords and lemmatize:**
- The NLTK WordNetLemmatizer is used to reduce words to their base forms, and stopwords (common words like "and", "the") are removed. This step involves splitting the text into words, processing each word, and then recombining them into a single string.

The preprocess_text function is applied to the text column of the DataFrame using the apply method, which transforms each entry in the column according to the defined preprocessing steps. The result is a DataFrame with a cleaned text column, ready for further analysis and model training.

```python
# Preprocessing Function
def preprocess_text(text):
    # Remove special characters, numbers, extra white spaces, and single characters
    text = re.sub(r'\W', ' ', str(text))
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = text.lower()
    # Remove stopwords and lemmatize
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    text = ' '.join(words)
    return text

df['text'] = df['text'].apply(preprocess_text)
df
```

### Naive Bayes Classifier (MultinomialNB)
This section details the process of training and evaluating a Naive Bayes classifier using a pipeline. The pipeline integrates text vectorization, TF-IDF transformation, and the Multinomial Naive Bayes model.

#### Pipeline Creation

- A pipeline (nb_pipeline) is created using Pipeline from sklearn.pipeline.
- Steps in the Pipeline:
    - CountVectorizer: Converts text data into a matrix of token counts.
    - TfidfTransformer: Transforms the count matrix to a normalized TF-IDF representation.
    - MultinomialNB: Applies the Multinomial Naive Bayes classifier on the transformed data.

#### Training the Naive Bayes Model:

- The pipeline is trained using the fit method on the training data (X_train, y_train).

#### Predictions:

- Predictions on the test set (X_test) are made using the predict method of the trained pipeline, resulting in y_pred_nb.

#### Evaluation Metrics:

- The accuracy of the model is calculated using accuracy_score from sklearn.metrics and printed.
- A detailed classification report is generated using classification_report from sklearn.metrics, providing precision, recall, F1-score, and support for each class.


```python
# Pipeline with CountVectorizer and MultinomialNB
nb_pipeline = Pipeline([
    ('vect', CountVectorizer()), # CountVectorizer for converting text to vectors
    ('tfidf', TfidfTransformer()), # TF-IDF transformation
    ('clf', MultinomialNB()), # Multinomial Naive Bayes classifier
])

# Train the Naive Bayes model
nb_pipeline.fit(X_train, y_train)

# Predictions on the test set
y_pred_nb = nb_pipeline.predict(X_test)

# Print the evaluation metrics (accuracy and classification report)
print("Naive Bayes Classifier Metrics:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_nb):.4f}")
print(classification_report(y_test, y_pred_nb))
```

The evaluation metrics indicate the performance of the Naive Bayes classifier. With an accuracy of approximately 93.76%, the model shows high precision, recall, and F1-scores for both classes. The metrics are consistent across the classes, suggesting that the model performs well in distinguishing between fake and real news.

### Word2Vec with LSTM (Deep Learning Model)

This section outlines the process of building and evaluating a deep learning model that combines Word2Vec embeddings with an LSTM network for fake news detection. Initially, the text data is tokenized into words with a maximum of 10,000 words and a maximum sequence length of 200. These sequences are then padded to ensure uniform length across all samples. The tokenized sequences are subsequently split into training and test sets using an 80-20 split.

A Word2Vec model is trained using the Gensim library to generate word embeddings from the text data. This involves splitting each text into words and training the Word2Vec model to produce 100-dimensional word vectors. An embedding matrix is created to map words to their corresponding Word2Vec vectors, including a padding token for sequences.

An LSTM model is defined using Keras, incorporating the pre-trained Word2Vec embeddings. The architecture includes an embedding layer initialized with the Word2Vec embeddings, an LSTM layer with 128 units, a dense layer with ReLU activation, a dropout layer, and a final dense layer with sigmoid activation for binary classification. The model is compiled with binary cross-entropy loss and the Adam optimizer, and it is trained for 10 epochs with a batch size of 128. The training progress is validated on the test set to monitor performance.

Predictions are made on the test set, and evaluation metrics such as accuracy, precision, recall, and F1-score are computed and displayed. The model achieves an impressive accuracy of 99.45%, with high precision, recall, and F1-scores for both classes. This indicates that the Word2Vec with LSTM model is highly effective in capturing the sequential nature and semantic meaning of the text, leading to excellent performance in fake news detection. This deep learning approach provides a significant improvement over simpler models, demonstrating the potential of advanced techniques in handling complex text classification tasks.

```python
# Tokenize and pad sequences
max_words = 10000  # Number of words to tokenize
maxlen = 200  # Max sequence length
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X)

# Convert texts to sequences
sequences = tokenizer.texts_to_sequences(X)
X_seq = pad_sequences(sequences, maxlen=maxlen)

# Splitting sequences into training and test sets
X_train_seq, X_test_seq, y_train_seq, y_test_seq = train_test_split(X_seq, y, test_size=0.2, random_state=42)

# Train Word2Vec model using Gensim
sentences = [text.split() for text in X]  # Split each text into words
embedding_dim = 100  # Dimension of word embeddings
w2v_model = Word2Vec(sentences, vector_size=embedding_dim, window=5, min_count=1, workers=4)

# Prepare the embedding matrix
vocab_size = min(max_words, len(tokenizer.word_index)) + 1  # Add 1 for padding token
embedding_matrix = np.zeros((vocab_size, embedding_dim))

for word, i in tokenizer.word_index.items():
    if i >= max_words:
        continue
    try:
        embedding_vector = w2v_model.wv[word]
        embedding_matrix[i] = embedding_vector
    except KeyError:
        # Word not in vocabulary, leave the row with zeros
        pass

# Define the number of LSTM units
lstm_units = 128  # Number of LSTM units

# Build LSTM model with pre-trained Word2Vec embeddings
model_lstm = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, weights=[embedding_matrix], input_length=maxlen, trainable=False),
    LSTM(lstm_units),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile model
model_lstm.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

# Model summary
print("Word2Vec with LSTM Model Architecture:")
print(model_lstm.summary())

# Train the model
epochs = 10  # Number of epochs
batch_size = 128  # Batch size

history = model_lstm.fit(X_train_seq, y_train_seq, epochs=epochs, batch_size=batch_size, validation_data=(X_test_seq, y_test_seq))

# Predictions
y_pred_prob = model_lstm.predict(X_test_seq)
y_pred_lstm = (y_pred_prob > 0.5).astype("int32")

# Print out evaluation metrics
print("\nWord2Vec with LSTM Model Metrics:")
print(f"Accuracy: {accuracy_score(y_test_seq, y_pred_lstm):.4f}")
print(classification_report(y_test_seq, y_pred_lstm))
```
